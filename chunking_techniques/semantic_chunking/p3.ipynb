```python
import re

# Split into sentences
single_sentences = re.split(r'(?<=[.?!])\s+', text)

print(f"{len(single_sentences)} sentences found")

sentences = [{"sentence": s, "index": i} for i, s in enumerate(single_sentences)]
sentences[:3]
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings_model = OpenAIEmbeddings(
    model="text-embedding-3-small",
    api_key=""
)

sentence_embeddings = embeddings_model.embed_documents(
    [s["sentence"] for s in sentences]
)

for i, emb in enumerate(sentence_embeddings):
    sentences[i]["embedding"] = emb

sentences[0]
```

```python
import json

# Save the embeddings to embeddings.json
with open("embeddings1.json", "w") as f:
    json.dump(sentence_embeddings, f)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

distances = []

for i in range(len(sentences) - 1):
    current_emb = sentences[i]["embedding"]
    next_emb = sentences[i + 1]["embedding"]

    similarity = cosine_similarity([current_emb], [next_emb])[0][0]
    distance = 1 - similarity

    distances.append(distance)
    sentences[i]["distance_to_next"] = distance

distances[:5]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
plt.plot(distances)
plt.title("Semantic Distance Between Consecutive Sentences")
plt.xlabel("Sentence Index")
plt.ylabel("Cosine Distance")

# Choose percentile threshold
percentile = 80
threshold = np.percentile(distances, percentile)

plt.axhline(y=threshold)
plt.show()

print("Breakpoint threshold:", threshold)
```

```python
# %%
breakpoints = [i for i, d in enumerate(distances) if d > threshold]

print("Breakpoint indices:", breakpoints)
print("Total chunks:", len(breakpoints) + 1)
```

```python
chunks = []
start_index = 0

for bp in breakpoints:
    chunk = " ".join([s["sentence"] for s in sentences[start_index:bp+1]])
    chunks.append(chunk)
    start_index = bp + 1

# Last chunk
if start_index < len(sentences):
    chunk = " ".join([s["sentence"] for s in sentences[start_index:]])
    chunks.append(chunk)

print(f"{len(chunks)} chunks created")
```

```python
# Preview chunks
for i, chunk in enumerate(chunks):
    print(f"\n--- Chunk {i+1} ---\n")
    print(chunk[:300])
```
