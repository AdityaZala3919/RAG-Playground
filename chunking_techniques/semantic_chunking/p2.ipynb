```python
import re

# Splitting the text on '.', '?', and '!'
single_sentences_list = re.split(r'(?<=[.?!])\s+', text)
print (f"{len(single_sentences_list)} senteneces were found")
```

```python
sentences = [{'sentence': x, 'index' : i} for i, x in enumerate(single_sentences_list)]
sentences[:]
```

```python
def combine_sentences(sentences, buffer_size=1):
    for i in range(len(sentences)):
        combined_sentence = ''
        for j in range(i - buffer_size, i):
            if j >= 0:
                combined_sentence += sentences[j]['sentence'] + ' '

        combined_sentence += sentences[i]['sentence']

        for j in range(i + 1, i + 1 + buffer_size):
            if j < len(sentences):
                combined_sentence += ' ' + sentences[j]['sentence']
        sentences[i]['combined_sentence'] = combined_sentence
    return sentences
```

```python
sentences = combine_sentences(sentences)
sentences[:3]
```

# Quick Revision: `combine_sentences` Function

- **Purpose:**  
  For each sentence in a list, create a new string that combines:
    - The sentence itself
    - A specified number (`buffer_size`) of sentences before and after it

- **How it works:**
    1. Loop through each sentence by index `i`.
    2. Initialize an empty string `combined_sentence`.
    3. Add up to `buffer_size` previous sentences (if they exist) to `combined_sentence`.
    4. Add the current sentence.
    5. Add up to `buffer_size` next sentences (if they exist).
    6. Store the result in a new key `'combined_sentence'` for each sentence dict.
    7. Return the modified list.

- **Parameters:**
    - `sentences`: List of dicts, each with a `'sentence'` key.
    - `buffer_size`: How many sentences before/after to include (default: 1).

- **Example:**
    - For `buffer_size=1`, each sentence is combined with its immediate neighbors.

- **Result:**  
  Each dict in `sentences` gets a new key `'combined_sentence'` with the combined text.

```python
from langchain_openai import OpenAIEmbeddings
oaiembeds = OpenAIEmbeddings(
    model="text-embedding-3-small",
    api_key=""
)
```

```python
embeddings = oaiembeds.embed_documents([x['combined_sentence'] for x in sentences])
```

```python
for i, sentence in enumerate(sentences):
    sentence['combined_sentence_embedding'] = embeddings[i]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_distances(sentences):
    distances = []
    for i in range(len(sentences) - 1):
        embedding_current = sentences[i]['combined_sentence_embedding']
        embedding_next = sentences[i + 1]['combined_sentence_embedding']
        
        similarity = cosine_similarity([embedding_current], [embedding_next])[0][0]        
        distance = 1 - similarity
        distances.append(distance)
        sentences[i]['distance_to_next'] = distance
    return distances, sentences
```

```python
distances, sentences = calculate_cosine_distances(sentences)
distances[:3]
```

# Quick Revision: `calculate_cosine_distances` Function

- **Purpose:**  
  Calculates cosine distances between consecutive sentence embeddings and stores the result.

- **How it works:**
    1. Loop through each sentence (except the last).
    2. For each pair of consecutive sentences:
        - Get their `'combined_sentence_embedding'`.
        - Compute cosine similarity using `cosine_similarity`.
        - Calculate distance as `1 - similarity`.
        - Append distance to `distances` list.
        - Store distance in current sentence dict as `'distance_to_next'`.
    3. Return the list of distances and the updated sentences list.

- **Parameters:**
    - `sentences`: List of dicts, each with a `'combined_sentence_embedding'` key.

- **Returns:**
    - `distances`: List of cosine distances between consecutive embeddings.
    - `sentences`: The input list, with each dict updated to include `'distance_to_next'`.

- **Note:**  
  Cosine distance = 1 - cosine similarity; measures how different two embeddings are.

```python
import numpy as np

plt.plot(distances)

y_upper_bound = .5
plt.ylim(0, y_upper_bound)
plt.xlim(0, len(distances))

breakpoint_percentile_threshold = 80
breakpoint_distance_threshold = np.percentile(distances, breakpoint_percentile_threshold) # If you want more chunks, lower the percentile cutoff
plt.axhline(y=breakpoint_distance_threshold, color='r', linestyle='-')

# Then we'll see how many distances are actually above this one
num_distances_above_theshold = len([x for x in distances if x > breakpoint_distance_threshold]) # The amount of distances above your threshold
plt.text(x=(len(distances)*.01), y=y_upper_bound/50, s=f"{num_distances_above_theshold + 1} Chunks")

# Then we'll get the index of the distances that are above the threshold. This will tell us where we should split our text
indices_above_thresh = [i for i, x in enumerate(distances) if x > breakpoint_distance_threshold] # The indices of those breakpoints on your list

# Start of the shading and text
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i, breakpoint_index in enumerate(indices_above_thresh):
    start_index = 0 if i == 0 else indices_above_thresh[i - 1]
    end_index = breakpoint_index if i < len(indices_above_thresh) - 1 else len(distances)

    plt.axvspan(start_index, end_index, facecolor=colors[i % len(colors)], alpha=0.25)
    plt.text(x=np.average([start_index, end_index]),
             y=breakpoint_distance_threshold + (y_upper_bound)/ 20,
             s=f"Chunk #{i}", horizontalalignment='center',
             rotation='vertical')
    
# # Additional step to shade from the last breakpoint to the end of the dataset
if indices_above_thresh:
    last_breakpoint = indices_above_thresh[-1]
    if last_breakpoint < len(distances):
        plt.axvspan(last_breakpoint, len(distances), facecolor=colors[len(indices_above_thresh) % len(colors)], alpha=0.25)
        plt.text(x=np.average([last_breakpoint, len(distances)]),
                 y=breakpoint_distance_threshold + (y_upper_bound)/ 20,
                 s=f"Chunk #{i+1}",
                 rotation='vertical')
        
plt.show()
```

```python
# Initialize the start index
start_index = 0

# Create a list to hold the grouped sentences
chunks = []

# Iterate through the breakpoints to slice the sentences
for index in indices_above_thresh:
    # The end index is the current breakpoint
    end_index = index

    # Slice the sentence_dicts from the current start index to the end index
    group = sentences[start_index:end_index + 1]
    combined_text = ' '.join([d['sentence'] for d in group])
    chunks.append(combined_text)
    
    # Update the start index for the next group
    start_index = index + 1

# The last group, if any sentences remain
if start_index < len(sentences):
    combined_text = ' '.join([d['sentence'] for d in sentences[start_index:]])
    chunks.append(combined_text)

# grouped_sentences now contains the chunked sentences
```

```python
for i, chunk in enumerate(chunks[:]):
    buffer = 200
    
    print (f"Chunk #{i}")
    print (chunk[:buffer].strip())
    print ("...")
    print (chunk[-buffer:].strip())
    print ("\n")
```
